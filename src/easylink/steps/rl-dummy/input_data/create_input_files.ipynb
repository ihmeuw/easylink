```python
import os
from pathlib import Path
import shutil

import numpy as np
import pandas as pd
import pseudopeople as psp
```

```python
file_stem_1 = "input_file_1"
file_stem_2 = "input_file_2"

psp_census = psp.generate_decennial_census().head(25)
df1 = psp_census
df2 = psp_census[0:20]  # df2 has most but not all of the records in df1
df1["Record ID"] = df1.index
df2["Record ID"] = df2.index

df1
```

```python
df2
```

```python
df1.to_parquet(f"{file_stem_1}.parquet")
df2.to_parquet(f"{file_stem_2}.parquet")
```

```python
known_clusters = pd.DataFrame(columns=["Input Record Dataset", "Input Record ID", "Cluster ID"])
known_clusters.to_parquet("known_clusters.parquet")
```

```python
# copy files to shared filesystem
output_dir = Path(f"/mnt/team/simulation_science/priv/engineering/er_ecosystem/input_data/{os.getlogin()}")
if not output_dir.exists():
    # make the directory with rwxrwxr-x permissions
    output_dir.mkdir(parents=True, mode=0o775)

for stem in [file_stem_1, file_stem_2, "known_clusters"]:
    filename = f"{stem}.parquet"
    filepath = output_dir / filename
    if filepath.exists():
        print(f"Overwriting existing file: {filepath}")
    shutil.copy(filename, filepath)
    print(f"File copied to: {filepath}")
```
